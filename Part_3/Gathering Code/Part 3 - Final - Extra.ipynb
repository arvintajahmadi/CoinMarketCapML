Let's implement a backtesting system to build more confidence in the model:

```python
pf = cf.copy()
```

```python
pf.drop(columns=['Date'], inplace = True)
```

```python
def backtest(data, start = 2114, step = 1):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        pca = PCA(n_components=3)
        Xtrain = train.drop(columns=['Target_XMR'])
        Xtrain = pca.fit_transform(Xtrain)
        ytrains = train['Target_XMR'] 
        Xtest = test.drop(columns=['Target_XMR'])
        Xtest = pca.transform(Xtest)
        scaler = StandardScaler()
        Xtrain = scaler.fit_transform(Xtrain)
        Xtest = scaler.transform(Xtest)
        models = RandomForestClassifier(n_estimators=2000, min_samples_split=20, random_state=1, bootstrap=True, oob_score=True, max_samples=1000)
        models.fit(Xtrain, ytrains)
        predictions = models.predict(Xtest)
        all_predictions.append(predictions)
    return all_predictions
```

Now we will try to use sentiment analysis on Bitcoin's wikipedia page to answer that question:

```python
import mwclient
import time
```

```python
site = mwclient.Site('en.wikipedia.org')
page = site.pages['Bitcoin']
```

```python
revs = list(page.revisions())
revs = sorted(revs, key = lambda rev: rev['timestamp'])
revs[0]
```

```python
from transformers import pipeline
sentiment_pipeline = pipeline('sentiment-analysis')
```

```python
def find_sentiment(text):
    sent = sentiment_pipeline([text[:250]])[0]
    score = sent['score']
    if sent['label'] == "NEGATIVE":
        score *= -1
    return score
```

```python
find_sentiment('I love you'), find_sentiment('I hate you')
```

```python
revs[1650]
```

```python
edits = {}

for rev in revs:
    date = time.strftime("%y-%m-%d", rev['timestamp'])
    if date not in edits:
        edits[date] = dict(sentiments = list(), edit_counts = 0)

    edits[date]['edit_counts'] += 1

    try:    
        comment = rev['comment']
    except:
        comment = 'nothing'
        
    edits[date]['sentiments'].append(find_sentiment(comment))
```

```python
for key in edits:
    if len(edits[key]['sentiments']) > 0:
        edits[key]['sentiment'] = np.mean(edits[key]['sentiments'])
        edits[key]['neg_sentiment'] = len([s for s in edits[key]['sentiments'] if s < 0]) / len(edits[key]['sentiments'])
    else:
        edits[key]['sentiment'] = 0
        edits[key]['neg_sentiment'] = 0

    del edits[key]['sentiments']
```

```python
def backtest(data, start = 3263, step = 1):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        pca = PCA(n_components=3)
        Xtrain = train.drop(columns=['Target_BTC'])
        Xtrain = pca.fit_transform(Xtrain)
        ytrains = train['Target_BTC'] 
        Xtest = test.drop(columns=['Target_BTC'])
        Xtest = pca.transform(Xtest)
        scaler = StandardScaler()
        Xtrain = scaler.fit_transform(Xtrain)
        Xtest = scaler.transform(Xtest)
        models = RandomForestClassifier(n_estimators=300, min_samples_split=20, random_state=1, bootstrap=True, oob_score=True, max_samples=1000)
        models.fit(Xtrain, ytrains)
        predictions = models.predict(Xtest)
        all_predictions.append(predictions)
    return all_predictions
```
